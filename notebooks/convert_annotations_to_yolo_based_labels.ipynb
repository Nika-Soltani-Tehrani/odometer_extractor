# Imports

```python
import os
import xml.etree.ElementTree as ET
```

# Data Path

```python
voc_annotations_path = '/content/drive/MyDrive/trodo-v01/trodo-v01/pascal voc 1.1/Annotations'
yolo_labels_path = '/content/drive/MyDrive/labels/'
images_path = '/content/drive/MyDrive/trodo-v01/trodo-v01/images/'
```

```python
os.makedirs(yolo_labels_path, exist_ok=True)
```

# Functions

```python
def convert_pascal_voc_to_yolo(voc_file, yolo_file, image_file, classes):
  tree = ET.parse(voc_file)
  root = tree.getroot()

  image_width = int(root.find('size/width').text)
  image_height = int(root.find('size/height').text)

  with open(yolo_file, 'w') as yolo_f:
    for obj in root.findall('object'):
      class_name = obj.find('name').text
      if class_name != 'odometer':
        continue

      class_id = classes.index(class_name)

      bndbox = obj.find('bndbox')
      xmin = float(bndbox.find('xmin').text)
      ymin = float(bndbox.find('ymin').text)
      xmax = float(bndbox.find('xmax').text)
      ymax = float(bndbox.find('ymax').text)

      # Convert to YOLO format
      x_center = (xmin + xmax) / 2.0 / image_width
      y_center = (ymin + ymax) / 2.0 / image_height
      width = (xmax - xmin) / image_width
      height = (ymax - ymin) / image_height

      yolo_f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```python
classes = ['odometer']

for voc_file in os.listdir(voc_annotations_path):
  if voc_file.endswith('.xml'):
    voc_file_path = os.path.join(voc_annotations_path, voc_file)
    yolo_file_path = os.path.join(yolo_labels_path, voc_file.replace('.xml', '.txt'))
    image_file_path = os.path.join(images_path, voc_file.replace('.xml', '.jpg'))
    convert_pascal_voc_to_yolo(voc_file_path, yolo_file_path, image_file_path, classes)

print("Conversion completed!")
```
